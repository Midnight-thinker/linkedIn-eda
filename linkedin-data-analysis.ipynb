# How to download the dataset?

LinkedIn provides access to various datasets through its LinkedIn Sales Navigator, LinkedIn Marketing Solutions, and LinkedIn Recruiter products. However, the process of downloading a dataset varies depending on the product you are using.

#### Here is a general overview of how to download a dataset from LinkedIn:

* Log in to your LinkedIn account and navigate to the profile button generally termed as 'Me' on the top right corner.

* Click on the 'Settings & Privacy' button.

* Tap onto the 'Data Privacy' option from the sidebar.

* Press the 'Get a copy of your data' to access the data of your own account.

* Identify the dataset you want to download and select it thereafter press the 'Request Archive' button .

* Depending on the amount of selection and your data a download link will be sent after several minute in your mail.

* Select a location on your computer where you want to save the downloaded file.

* Click the "Download" button to begin the download process.

* Once the download is complete, locate the downloaded file on your computer and open it using a spreadsheet or data analysis tool.

It's important to note that downloading datasets from LinkedIn may require a subscription to one of their premium products, and certain datasets may have usage restrictions. Additionally, ensure that you have the necessary permissions and comply with LinkedIn's terms of service when using any downloaded data.

### Extracting the time from Date column's datetime datatype:

The `strftime()` function in pandas is used to format datetime objects in a specific format. It returns a string representing the datetime object in the specified format.

The `strftime()` function is a method of the pandas DatetimeIndex object and can be accessed using the dt accessor. Thus we're using `strftime()` to extract time from Date column's datetime datatype:

```python
messages_df['TIME'] = messages_df['DATE'].dt.strftime("%H:%M:%S")
```

## Q1) At what time most invitations sent?

```python
most_invitations_sent=invitation_df[invitation_df['From']==user].groupby(['Time']).size().sort_index(ascending=False)
most_invitations_sent
```

Here we checking at what time most of the invitations were sent by the user using `size()` function, `groupby()`function and `sort_index()` function.

This particular line graph provides an idea about when most of the invitations were sent which is 09:29:00. The y-axis indicates the number of invitations sent and the x-axis indicates when those invitations were sent.

## Q7) Which contacts hold the maximum incoming messages?

```python
hieghest_msg_incoming=messages_df[messages_df['FROM']!= user].groupby("FROM")["CONTENT"].count().sort_values(ascending=False)
hieghest_msg_incoming
msg_incoming= hieghest_msg_incoming.to_frame()
msg_incoming
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title( "Maximum incoming messages" , size = 15 )
fig.patch.set_facecolor('white')
s = sns.histplot(data=msg_incoming,x=msg_incoming.CONTENT)
ax.set_xticklabels(msg_incoming.index, rotation=45, ha='right')
ax.set_xlabel( "From" , size = 10)
ax.set_ylabel( "Count" , size = 10)
```

The given histplot showing the contact who sent maximum number messages to the user. The person with highest incoming messages was 'Ritesh Gupta'. The y-axis shows count of total incoming messages and x-axis shows the person who sent those highest number of messages.

## Q8) What is the percentage of total messages received?

```python
dms_recieved_total = messages_df[messages_df['FROM']!= user].groupby("FROM")["CONTENT"].count().sort_values(ascending=False).sum()
dms_recieved_total
```

#### Total number of messages received are: 346 

```python
dms_sent_total = messages_df[messages_df['FROM']== user].groupby("FROM")["CONTENT"].count().sort_values(ascending=False).sum()
dms_sent_total
```

#### Total number of messages sent are: 164 

```python
total_dms = dms_recieved_total + dms_sent_total
total_dms
```

#### Total number of messages sent and received are: 510 

```python
percentage_sent_dms = (dms_sent_total / total_dms) * 100
round(percentage_sent_dms,2)
```

#### Percentage of total sent messages are: 32.16% 

## Q9) What is the percentage of total messages sent?

```python
percentage_recieved_dms = (dms_recieved_total / total_dms) * 100
round(percentage_recieved_dms,2)
```

#### Percentage of total recieved messages are: 67.84% 

```python
messages_df['YEAR'] = messages_df['DATE'].dt.year
years= messages_df['YEAR'].to_list()
```

```python
plt.bar(years,percentage_sent_dms);
plt.bar(years,percentage_recieved_dms,bottom= percentage_sent_dms);
plt.title("Ratio of Total Sent & Recieved Message")
plt.xlabel('Time', fontweight='bold');
plt.ylabel('Ratio', fontweight='bold');
```

The above barplot is presenting a ratio between number of Total sent & received messages. The blue portion indicates the percentage of total sent messages and the orange portion indicates the total percentage of recieved messages.

## Q12) At which weekday usage of Linkedin by the user is highest?

In this code we're going to check in which day usage of linkedIn was highest. That is in which most of the messages were sent or recieved.

```python
messages_df['WEEKDAY'] = messages_df['DATE'].dt.dayofweek
ans=messages_df['WEEKDAY'].max()
if ans == 0:
        print("Monday")
elif ans == 1:
        print("Tuesday")
elif ans == 2:
        print("Wednesday")
elif ans == 3:
        print("Thursday")
elif ans == 4:
        print("Friday")
elif ans == 5:
        print("Saturday")
else:
        print("Sunday")
```
